# OrdinalEncoder with loop

```python
df15.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder
from pandas.core.dtypes.common import is_numeric_dtype
```

```python
cols = df15.columns
cols
```

```python
for ord in cols:
    if is_numeric_dtype(df15[ord]):
        continue
        
    else:
        ord_encoer = OrdinalEncoder(categories=[df15[ord].unique()])
        enc_values = ord_encoer.fit_transform(df15[[ord]])
        enc_values = pd.DataFrame(enc_values,columns=[ord])
        df15 = pd.concat([df9.drop(ord,axis=1),enc_values],axis=1)
df9.head()
```
