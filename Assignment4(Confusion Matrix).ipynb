```python
x = df.drop('DEATH_EVENT', axis='columns')
```

```python
x.head()
```

# X,Y Seperate

```python
y = df['DEATH_EVENT']
```

```python
y.head()
```

# Split DataSet 

```python
xtrain, xtest, ytrain, ytest = tts(x,y,test_size=.30,random_state=1)
```

```python
xtrain.head()
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
clf = DecisionTreeClassifier() #creating decision tree
```

```python
clf.fit(xtrain,ytrain)
```

```python
from sklearn import tree
plt.figure(figsize=(16,14))
tree.plot_tree(clf,filled=True,rounded=True,feature_names=x.columns)
```

```python
pred = clf.predict(xtest)
```

```python
pred #prdict
```

```python
np.array(ytest) #actual
```

# Confusion Matrix

```python
classification_report(ytest,pred)
```

```python
print(classification_report(ytest,pred))
```

```python
cm = confusion_matrix(ytest,pred)
```

```python
cm
```

```python
sns.heatmap(cm, annot=True)
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')
plt.title('Confusion Matrix')
```

```python
tp, fn, fp, tn = confusion_matrix(ytest, pred).reshape(-1)
```

```python
tp
```

```python
fn
```

```python
fp
```

```python
tn
```

# Accurecy 

```python
accurecy = (tp+tn)/(tp+fp+tn+fn)
accurecy
```

# Presion PPV

```python
pre = tp/(tp+fp)
pre
```

# True Positive Rate or Sensitivity

```python
TPR = tp/(tp+fn)
TPR
```

# False Positive Rate or 1 - Specificity 

```python
FPR = fp/(fp+tn)
FPR
```

# Recall


```python
Recall = tp/(tp+fn)
Recall
```

# F1 Score

```python
# F1 = (precision + Recall)/2
F1 = ( 2 * pre * Recall)/(pre + Recall)
F1
```

# Specificity, Selectivity or True Negative Rate

```python
# Specificity = True Negative / (True Negative + False Positive) 
specificity = tn/(tn+fp)
specificity
```

# Threat Score (TS) or Critical Success Index (CSI)

```python
# CSI = TP/(TP + FN + FP)
CSI = tp/(tp+fn+fp)
CSI
```

# False Discovery Rate

```python
# FDR = FP/(FP+TP)
FDR = fp/(fp+tp)
FDR
```

```python
from sklearn.metrics import matthews_corrcoef
matthews_corrcoef(ytest,pred)
```
